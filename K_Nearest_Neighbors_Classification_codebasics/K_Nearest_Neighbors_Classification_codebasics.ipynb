**Theory behind KNN classifier**
1) First, you set the K value of your KNN model (you should figure out the best K value based on your problem). Let's say for here, you have built a KNN model with K=3, then you need to use the model to classify the new (yellow) sample. This means the model will figure out the nearest 3 samples/data points (by calculating the distance [usually is euclidean distance] between the new sample and each of the nearest samples) and use them to classify (predict the class) the new sample. By looking at the graph itself, you immediately get an intuition that this new sample has to be virginica (blue color) because it is nearer to this virginica cluster/group, and KNN model works just like that.   <br />
    <img src="hidden\photo1.png" alt="This image describes the theory behind KNN model, Part 1" style="width: 400px;"/>  <br />

2) By using the euclidean distance, the nearest 3 samples to the new sample are these 3 circled samples.    <br />
    <img src="hidden\photo2.png" alt="This image describes the theory behind KNN model, Part 2" style="width: 400px;"/>  <br />

3) Since these 3 circled samples belong to the class of virginica (blue color), the KNN model will classify the new sample as virginica (blue color).   <br />
    <img src="hidden\photo3.png" alt="This image describes the theory behind KNN model, Part 3" style="width: 400px;"/>  <br />

4) Let's say, if the same model is built with K=3. The model will figure out 10 nearest samples from the new sample. So here, you have 7 virginica and 2 versicolor samples circled. Since the highest number among the circled (K) samples belong to virginica, the model will classify the new sample as virginica.   <br />
    <img src="hidden\photo4.png" alt="This image describes the theory behind KNN model, Part 4" style="width: 400px;"/>  <br />

5) Let's say,  if the same model is built with K=20, you will have a problem. Because in the dataset, the total number of sample in virginica class are very less. The KNN model is now misclassifying the new sample by classifying it as versicolor instead of virginica. We know this new sample is misclassified intuitively, because clearly by looking at the graph, you know that this new sample is closer to the virginica samples. Hence, you have to carefully choose the K value such that it is not very high or very small, depends on your dataset.   <br />
    <img src="hidden\photo5.png" alt="This image describes the theory behind KNN model, Part 5" style="width: 400px;"/>  <br />

6) One thing I would like to clarify is that here we had only two features but KNN works equally if there are more than two features. There could be N number of features and KNN will just work fine. I mean as a human we can't visualize it, but mathematically it should just work okay.   <br />

7) Rule of KNN classifier: The KNN model will figure out the K nearest samples, then take the maximum count.   <br />

# Load the dataset

```python
import pandas as pd

# import the sklearn ready-made dataset of iris
from sklearn.datasets import load_iris

# load the dataset into the variable called iris
iris = load_iris()

# show the attributes of the dataset
print(dir(iris))

'''
The independent variable names are available in the attribute of feature_names of the dataset (sepal length (cm),sepal width (cm),petal length (cm),petal width (cm))
The data of the independent variables are available in the attribute of data of the dataset
The data of the dependent variable is available in the attribute of target of the dataset (0 for setosa, 1 for versicolor, 2 for virginica)
The representation of data of the dependent variable are available in the attribute of target_names of the dataset (setosa, versicolor, virginica) [Extra information: each of them is a species/type of iris flower]
'''
```

```python
# show the data available in the attribute of feature_names of the dataset
print(iris.feature_names)
```

```python
# create a dataframe called df, by using the data available in the attribute of data of the dataset (data of independent variables),
# and use the data available in the attribute of feature_names of the dataset as the dataframe column names.
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# show the df dataframe
df
```

```python
# show the data available in the attribute of target_names of the dataset
print(iris.target_names)
```

```python
# create a new column called target on the df dataframe, which stores the data of dependent variable (data available in the attribute of target of the dataset)
df['target'] = iris.target

# show the df dataframe
df
```

### Data exploration

This list will be used as the parameter of labels of the confusion matrix at later stage

```python
import pandas as pd
# create a simple pandas series from a list of target of the dataset (under digits)
categorical_label_list = pd.Series(iris.target)
# get the name of all labels/categories available in the target of the dataset (under digits)
name_categorical_label_list = categorical_label_list.value_counts().index
# sort the name of all labels/categories available in the target of the dataset (under digits) in ascending order
name_categorical_label_list_sorted = name_categorical_label_list.sort_values(ascending=True)
# show the sorted name of all labels/categories available in the target of the dataset (under digits) in ascending order
print(name_categorical_label_list_sorted)
```

#### Uses the data of a specified column to create another column, with apply()

```python
# the function uses the data of a specified column to create another column.
# the function gets each data of the target column as x, then the data in the specified variable indexed by the same x is obtained.
# the function then transforms each of the data of x into the data in the specified variable indexed by the same x.
# after transforming all data available, all the transformed data are made into a new column.
# Here, we named the new column as flower_name. Then, concatenate the new column to the df dataframe.
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])

# show the df dataframe
df
```

### Data preprocessing

#### Get the independent variables and dependent variable of the dataset

```python
# load the data of independent variables available in df dataframe to variable X (drop the remaining data of df dataframe). We drop the columns of target and flower_name
X = df.drop(['target','flower_name'], axis='columns')
X.head()
```

```python
# load the data of dependent variable available in df dataframe to variable Y
Y = df.target
Y.head()
```

#### Split dataset into training and testing data using train_test_split() method

```python
from sklearn.model_selection import train_test_split

# split 20% of the dataset into test set, 80% of the dataset into train set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)

# check the number of samples in the train set and test set
print('The train set consists of '+ str(len(X_train)) + ' sample(s) and the test set consists of '+ str(len(X_test)) + ' sample(s)')
```

### Develop machine learning model (K Nearest Neighbors (KNN) Classifier/Model)

```python
from sklearn.neighbors import KNeighborsClassifier

# create the KNN classfier/model, with K=10 (consider the 10 nearest samples/data points) & using Euclidean distance as distance metric
knn = KNeighborsClassifier(n_neighbors=10)

# train the model
knn.fit(X_train,Y_train)

# compute the accuracy of the trained model
print('The accuracy of the trained KNN classifier: ',knn.score(X_test,Y_test))
```

### Evaluate the performance of the trained machine learning model (KNN model)

#### Compute the confusion matrix 

Evaluate the confusion matrix (To see the distribution of errors [where the model performed well vs poorly])

```python
# use the trained model to make prediction over the samples in X_test
Y_predicted = knn.predict(X_test) 

from sklearn.metrics import confusion_matrix

'''
compute/generate the confusion matrix of the trained model using: 
1) the samples in Y_test (ground truths in test set), 
2) the predictions over the samples in X_test (features in test set),
3) the name of all labels/categories available in the dataset as a list in ascending order
'''
cm = confusion_matrix(Y_test, Y_predicted, labels=name_categorical_label_list_sorted) 
# show the confusion matrix
print(cm)
```

##### Visualize the confusion matrix

1) The value shown in each diagonal cell represents the number of correct prediction made to the corresponding class in the test set.   <br \>
2) The value shown in each non-diagonal cell represents the number of wrong prediction made to the corresponding class in the test set.   <br \>

```python
# code for confusion matrix visualization
import seaborn as sn

plt.figure(figsize = (10,7)) # generate a figure window
sn.heatmap(cm, annot=True) # generate a heatmap with the content of variable cm (the confusion matrix). annot=True means show the colorbar beside the heat map
plt.xlabel('Predicted', fontsize=13) # label the x-axis of the heatmap
plt.ylabel('Ground truth', fontsize=13) # label the y-axis of the heatmap
plt.title('Confusion Matrix',fontsize=17) # label the title of the heatmap
```

#### Generate the classification report

```python
from sklearn.metrics import classification_report

'''
generate the classification report of the trained model using: 
1) the samples in Y_test (ground truths in test set), 
2) the predictions over the samples in X_test (features in test set),
3) the name of all labels/categories available in the dataset as a list in ascending order
'''
cr = classification_report(Y_test, Y_predicted, labels=name_categorical_label_list_sorted)
# show the classification report
print(cr)
```
