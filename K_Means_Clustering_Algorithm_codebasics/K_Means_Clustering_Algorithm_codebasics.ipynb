Using this dataset, we try to identify the underlying structure in the dataset (EG: we sometimes try to find the clusters in that dataset), and we can make useful predictions out of it. K means is a very popular clustering algorithm and it is an unsupervised machine learning algorithm.  <br />
   2) Let's say you have a dataset as below, where x and y axes represent 2 different independent variables (features) respectively, and you want to identify clusters in this dataset. Now when the dataset is given to you, you don't have any information on dependent/target variables, so you don't know what you are looking for.  <br />

K in K means clustering algorithm is a free parameter, it represents the number of cluster centroids OR the number of clusters/groups required (you want) to identify from the dataset. Wherein, before you start the algorithm you have to tell the algorithm what is the value of k that you are looking for. Here, our k equals to 2 <br />

The simple mathematical way to identify the distance is to draw a line between the centroids and then draw a perpendicular line. Anything on the left hand side is Cluster 1 (red color), anything on the right hand side is Cluster 2 (green color). Now, you already have your 2 imperfect clunky clusters.  <br />

3) **Theory behind Elbow method:**  <br />
   1) You need to supply k value to your K means clustering algorithm but what is a good number on k? Because here we have 2 dimensional space (features) but in reality, you will have so many features and it is hard to visualize that data on a scatter plot.  <br />
   2) Let's look at our dataset, we started with 2 clusters.  <br />
      <img src="hidden\explanation10.png" alt="This image describes explanation, part 11" style="width: 400px;"/>  <br />

   3) But someone might say no, these are actually 4 clusters.  <br />
      <img src="hidden\explanation11.png" alt="This image describes explanation, part 12" style="width: 400px;"/>  <br />

   4) Third person might say they are actually 6 clusters.  <br />
      <img src="hidden\explanation12.png" alt="This image describes explanation, part 13" style="width: 400px;"/>  <br />

   5) So you can see different people might interpret these things in a different way, and your job is to find out the best possible k number. That technique is called elbow method.  <br />

   6) Elbow method works by starting with a k value. Let's say we start with k=2, we try to compute the sum of square error (SSE). What it means is for each of the clusters, you try to compute the distance of individual data point from the centroid it belongs to, then you square it and then sum it up. So you have SSE 1 for Cluster 1 and SSE 2 for Cluster 2. You do this for all your clusters. By summing up the SSE of all your clusters, you get the total SSE. We do square just to handle negate value. So now we compute the SSE for K=2, you repeat the same process for k=3,4,...  <br />
      <img src="hidden\explanation13.png" alt="This image describes explanation, part 14" style="width: 400px;"/>  <br />

   7) Once you have SSE for different k values, you draw a plot as depicted below. You will realize that as you increase k value, the SSE decreases. It is kind of intuitive if you think about it at some point. You can consider all your data points as one cluster individual, where your SSE becomes almost zero. Let's assume we have only 11 data points and SSE will beomce 0 at k=11.  <br />

   8) The general guideline is to find out an elbow. On this chart, the point at k=4 is short of like an elbow. Hence, k=4 is a good cluster number.  <br />

# Load the dataset

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

# load the dataset from a CSV file to a variable called df in the format of dataframe.
df = pd.read_csv("income.csv")
# show the first 5 rows of the df dataframe
print(df)
```

# Data visualization

Visualize the distribution of the data in the dataset

```python
# plot the scatter plot of data: age vs income
plt.scatter(df['Age'],df['Income($)'])
plt.title('Data visualization: Age vs Income($)')
plt.xlabel('Age')
plt.ylabel('Income($)')
```

# Develop K means clustering algorithm without the scaled features of the dataset

K = n_clusters = Number of cluster centroids = Number of clusters/groups required to identify from the dataset

```python
from sklearn.cluster import KMeans
# create a KMeans object called km variable, with K=3 (n_clusters=3=Number of cluster centroids)
# Here, we set n_clusters=3 because we can intuitively realize there are 3 clusters/groups of samples/data points available on the scatter plot above
km = KMeans(n_clusters=3)
```

## Train the unsupervised machine learning model (K means clustering algorithm) without the scaled features and use the trained model to make predictions over the data used to train it

The unsupervised machine learning model is trained with the independent variables [features] only (unlike supervised machine learning model, it is trained with both independent and dependent variables), then it makes prediction [cluster] over each sample's features (unlike supervised machine learning model, it predicts the target/label of each sample based on the sample's independent variables)

```python
'''
run (train) K means algorithm on the features (age and income data) of the dataset available in the df dataframe, 
then the algorithm compute (predict) the cluster (which data point 
belongs to which cluster) based on the number of clusters we want to 
identify (n_clusters). fit_predict returns the cluster number (prediction of the trained model) that 
each sample of the dataset belongs to. For example, 1st entry of 
y_predicted represents the cluster number that the sample represented 
by 1st row of the dataset belongs to.
'''
y_predicted = km.fit_predict(df[['Age','Income($)']])

# show the predicted cluster/group number of each sample of the dataset
print('The predicted cluster/group number of Sample (Data point) 1 to the last sample:\n ',y_predicted)

# print the coordinates of the centroid of each cluster (1st row for the centroid of cluster 1, ...)
print('The coordinates of the centroid of Cluster 1 to the last cluster:\n ',km.cluster_centers_)
```

## Results visualization

```python
'''
create a new column called cluster on the df dataframe. 
Then, load the data of y_predicted variable (prediction of the trained model) to the newly created column. 
So now we can visualize which sample belongs to which cluster/group better.
'''
df['cluster'] = y_predicted

# show the first 5 rows of df dataframe
df.head()
```

```python
# create 3 different dataframes, each dataframe stores the samples (data points) belongs to 1 cluster

# create a dataframe called df1, to store all the rows from df dataframe (samples / data points) which cluster is 0
df1=df[df.cluster==0]
# create a dataframe called df2, to store all the rows from df dataframe (samples / data points) which cluster is 1
df2=df[df.cluster==1]
# create a dataframe called df3, to store all the rows from df dataframe (samples / data points) which cluster is 2
df3=df[df.cluster==2]

# plot the samples in all these 3 dataframes onto the same scatter plot for visualization
plt.scatter(df1.Age,df1['Income($)'],label='Cluster1',color='green')
plt.scatter(df2.Age,df2['Income($)'],label='Cluster2',color='red')
plt.scatter(df3.Age,df3['Income($)'],label='Cluster3',color='black')
plt.title('Data visualization: Age vs Income($)')
plt.xlabel('Age')
plt.ylabel('Income($)')
# plot the centroid/center of each cluster 
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],label='centroid_coordinates',color='purple')
plt.legend()

'''
There is problem for the samples / data points of particular clusters (because the samples are predicted as the member of the same group but their respective locations are not close to each other such that they don't look like a cluster/group)
If the scale of features of the dataset used to train a unsupervised machine learning model is not right (EG: If the K means algorithm model is trained with only 2 features, the scale of the feature represented by x-axis [Here, ranging from 27.5 to 42.5] is different with the scale of the feature represented by y-axis [Here, ranging from 40000 to 160000]), the model cannot group the samples/data points of the dataset correctly.
Hence, we must scale the features of the dataset (which used to train the K means algorithm) before training the model and use the model to make predictions, so that the model can group (predict) the samples/data points of the dataset correctly.
'''
```

# Scale each feature of the dataset using MinMaxScaler

1) If the scale of features of the dataset used to train a unsupervised machine learning model (K means algorithm model) is not correct (EG: If the K means algorithm model is trained with only 2 features, scale of the feature represented by x-axis is different with the scale of the feature represented by y-axis), the K means algorithm model cannot cluster/group the samples/data points of the dataset correctly. 
2) Hence, we must scale the features of the dataset (which used to train the K means algorithm) [Here, are age & income] before training the model and use the model to make predictions, so that the model can group (predict) the samples/data points of the dataset correctly.

```python
# Create a new dataframe called df_scaledfeatures. Load the dataset available in the CSV file into df_scaledfeatures dataframe. Later, update the scaled features to df_scaledfeatures dataframe.
df_scaledfeatures = pd.read_csv("income.csv")

from sklearn.preprocessing import MinMaxScaler

# create a MinMaxScaler object called scaler variable
scaler = MinMaxScaler()

## make the scale of the feature available in df_scaledfeatures dataframe called income into the range from 0 to 1 (so after the scaling for the income feature is done, you will have a scale of 0 to 1 on y-axis)
# with fit(), the MinMaxScaler learns how to scale the feature of the dataset (available in df_scaledfeatures dataframe) called income, by calculating the mean and standard deviation for each sample's income
scaler.fit(df_scaledfeatures[['Income($)']])

# with transform(), the MinMaxScaler scales (transforms) the data (each entry of the feature column called income, available in df_scaledfeatures dataframe) into a more suitable form (in the range from 0 to 1), according to the mean and standard deviation calculated during the fit() step. The scaled data replaces the original data of the income column of the df_scaledfeatures dataframe 
df_scaledfeatures['Income($)'] = scaler.transform(df_scaledfeatures[['Income($)']])

# show the _scaledfeatures dataframe
df_scaledfeatures
```

```python
## make the scale of the feature available in df_scaledfeatures dataframe called age into the range from 0 to 1 (so after the scaling for the age feature is done, you will have a scale of 0 to 1 on x-axis)
# with fit(), the MinMaxScaler learns how to scale the feature of the dataset (available in df_scaledfeatures dataframe) called age, by calculating the mean and standard deviation for each sample's age
scaler.fit(df_scaledfeatures[['Age']])

# with transform(), the MinMaxScaler scales (transforms) the data (each entry of the feature column called age, available in df_scaledfeatures dataframe) into a more suitable form (in the range from 0 to 1), according to the mean and standard deviation calculated during the fit() step. The scaled data replaces the original data of the age column of the df_scaledfeatures dataframe 
df_scaledfeatures.Age = scaler.transform(df_scaledfeatures[['Age']])

# show the _scaledfeatures dataframe
df_scaledfeatures
```

# Develop K means clustering algorithm with scaled features of the dataset

## Train the unsupervised machine learning model (K means clustering algorithm) with the scaled features and use the trained model to make predictions over the data used to train it

```python
# create a KMeans object called km variable, with K=3 (n_clusters=3=Number of cluster centroids)
# Here, we set n_clusters=3 because we can intuitively realize there are 3 clusters/groups of samples/data points available on the scatter plot above
km_scaledfeatures = KMeans(n_clusters=3)
'''
run (train) K means algorithm on the features (age and income data) of the dataset available in the df_scaledfeatures dataframe, 
then the algorithm compute (predict) the cluster (which data point 
belongs to which cluster) based on the number of clusters we want to 
identify (n_clusters). fit_predict returns the cluster number that 
each sample of the dataset belongs to. For example, 1st entry of 
y_predicted represents the cluster number (prediction of the trained model) that the sample represented 
by 1st row of the dataset belongs to.
'''
y_predicted_scaledfeatures = km_scaledfeatures.fit_predict(df_scaledfeatures[['Age','Income($)']])

# show the predicted cluster/group number of each sample of the dataset
print('The predicted cluster/group number of Sample (Data point) 1 to the last sample:\n ',y_predicted_scaledfeatures)

# print the coordinates of the centroid of each cluster (1st row for the centroid of cluster 1, ...)
print('The coordinates of the centroid of Cluster 1 to the last cluster:\n ',km_scaledfeatures.cluster_centers_)
```

## Result visualization

```python
'''
create a new column called cluster_scaledfeatures on the df_scaledfeatures dataframe. 
Then, load the data of y_predicted_scaledfeatures variable (prediction of the trained model) to the newly created column. 
So now we can visualize which sample belongs to which cluster/group better.
'''
df_scaledfeatures['cluster_scaledfeatures'] = y_predicted_scaledfeatures

# show the first 5 rows of df_scaledfeatures dataframe
df_scaledfeatures.head()
```

```python
# create 3 different dataframes, each dataframe stores the samples (data points) belongs to 1 cluster

# create a dataframe called df1_scaledfeatures, to store all the rows from df_scaledfeatures dataframe (samples / data points) which cluster is 0
df1_scaledfeatures=df_scaledfeatures[df_scaledfeatures.cluster_scaledfeatures==0]
# create a dataframe called df2_scaledfeatures, to store all the rows from df_scaledfeatures dataframe (samples / data points) which cluster is 1
df2_scaledfeatures=df_scaledfeatures[df_scaledfeatures.cluster_scaledfeatures==1]
# create a dataframe called df3_scaledfeatures, to store all the rows from df_scaledfeatures dataframe (samples / data points) which cluster is 2
df3_scaledfeatures=df_scaledfeatures[df_scaledfeatures.cluster_scaledfeatures==2]

# plot the samples in all these 3 dataframes onto the same scatter plot for visualization
plt.scatter(df1_scaledfeatures.Age,df1_scaledfeatures['Income($)'],label='Cluster1_scaledfeatures',color='green')
plt.scatter(df2_scaledfeatures.Age,df2_scaledfeatures['Income($)'],label='Cluster2_scaledfeatures',color='red')
plt.scatter(df3_scaledfeatures.Age,df3_scaledfeatures['Income($)'],label='Cluster3_scaledfeatures',color='black')
plt.title('Data visualization: Age vs Income($)')
plt.xlabel('Age')
plt.ylabel('Income($)')
# plot the centroid/center of each cluster 
plt.scatter(km_scaledfeatures.cluster_centers_[:,0],km_scaledfeatures.cluster_centers_[:,1],label='centroid_coordinates',color='purple')
plt.legend() # we have parameter label in each plt.scatter code so that the labels will be included by plt.legend
```

# Elbow method

The codes below use the elbow method to determine the best k value for K means algorithm model, based on the features of the dataset which used to train the K means algorithm model

```python
# create a range of k=n_clusters value
k_rng = range (1,10)
# create an empty array called sse, to store the sum of square error (SSE) of the K means clustering algorithm at different k values
sse = []

for k in k_rng:
    # create a K means clustering algorithm
    km_ElbowMethod = KMeans(n_clusters=k)
    # train the algorithm with the scaled features of the dataset
    km_ElbowMethod.fit(df_scaledfeatures[['Age','Income($)']])
    # inertia will return the sum of square error (SSE) of the K means clustering algorithm at a k value
    sse.append(km_ElbowMethod.inertia_)

# show the sum of square error (SSE) of the K means clustering algorithm at different k values
print('The sum of square error (SSE) of the K means clustering algorithm of k=1 to the last k value:\n ',sse)

# plot the elbow plot to identify which k value represents the elbow
plt.plot(k_rng,sse)
plt.xlabel('K')
plt.ylabel('Sum of square error (SSE)')
```
